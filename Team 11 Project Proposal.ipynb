## Questions to answer 


- Motivation: 

People travels a lot nowadays, and it is quite often that we end up paying more than the person sitting next to us which could be pretty annoying and costly. Since flight price is determined on a dynamic basis, it is almost impossible to get to know the fluctuation of the ticket price. Therefore, we are thinking of using data visualization to plot out precious information to get to know the pattern of it and to find out the influencing factors behind it.

The biggest problem for our dataset is the size, it is around 30G which make bring us nightmare if we would like to load the entire dataset into our machine. Thus, we might have to trim the data either timewise or flightwise.

The potential surprising insight that we might find out could be that some routes might favor certain airlines so they can sell ticket at a higher price, close to monopoly, or, flight price might be cheaper in different days in a week.

- Related Work: 

In December 2022, there was an article published by NBC News describing the current situation of rising airline prices and high possibilities of flight cancellation. In the article, it mentioned a tweet which was published by the official American Airlines saying that the fares were raised up for some destinations. This tweet was followed by another tweet complaining the extremely high prices of $7,719 for just two economy tickets from Chicago to Washington D.C.

After read through the news, we started to think about several factors that may influence prices of airline tickets. Some factors we already know are seasons, events, vacations, etc. All these are related to dates. However, the above article brought our attention to other factors, such as different destination airports, different rows in the cabin, as well as different days in a week. Therefore, in this project, we would like to explore how these factors can affect the airline prices differently.  

Link to the article: https://www.nbcnews.com/business/consumer/airlines-show-high-ticket-prices-amid-southwest-flight-cancellations-rcna63510

*   Data

The dataset is a single .csv file containing every purchasable ticket found on Expedia between 2022-04-16 and 2022-10-05, to/from the following airports: ATL, DFW, DEN, ORD, LAX, CLT, MIA, JFK, EWR, SFO, DTW, BOS, PHL, LGA, IAD, OAK.

It has **5999739 rows** in total, and each row represents a flight. Each flight is described by **27 features**. 

The dataset is uploded to both Kaggle and Github 4 months ago. On both websites, although being downloaded almost 2000 times on Kaggle, currently there aren't any record indicating that this dataset is used in other works or projects. 


Link to the dataset: https://www.kaggle.com/datasets/dilwong/flightprices



*   Questions
1. What is the price difference for the same flight departing on different dates, and more importantly for the same flight but searched for on different days of the week? 
  
  -> This may serve as valuable information for discussing a good strategy for searching for and purchasing flight tickets in the future.
  
  -> This could be done by getting the day in a week of all searching and purchasing date data and grouping them to calculate the mean price of each day.

2.  Is the rate of seat remaining has anything to do with the airline, destination, or duration of the flight? If yes, then what kind of reasons may lead to these vacancies, and will the same flight price become cheaper for the following flights when there was some excess supply last time?

 -> This is also helpful information for us customers to have a better idea of flights of certain airlines or routes that might experience relatively strong fluctuations in their flight price so that we can have a better decision in terms of airline-wise selection or destination-wise selection.

 -> Grouping by airlines, destinations, and duration of flights and calculating the relative mean could give us a brief understanding of the relationship, then we could probably dive into those with odd values and look into the flight price in terms of the specific airline, destination, and duration of flights. 

3. Is there a difference in price for flights that have the same origin but different destination airports in the same city (e.g: from SFO - JFK, SFO - EWR)?
    
    -> This can be an interesting question because different destination airports have different distances away from the origin. Each airport in the same city has its own pros and cons. Some might be closer to the city center, while others might be closer to neighboring areas. Thus, prices can vary. Understanding this question can help people choose their destination airports better and help them save money. 

    -> First we need to group airports by their cities. For flights that fly to the same city but different airports, we can then group again by their origins. Then we can compare their prices.
